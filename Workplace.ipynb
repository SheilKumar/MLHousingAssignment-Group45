```python
import pandas as pd
import geopy
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="my_request")
df = pd.read_csv('House.csv')
df2 = pd.DataFrame(columns=["Price", "Address_x","Address_y", "Bedrooms", "Baths","Area"])

for i in range(0,df.shape[0]):
    if df.loc[i,'Address'] != None:
        location = geolocator.geocode(df.loc[i,'Address'])
        if location:
            cardAddress_x = location.latitude
            cardAddress_y = location.longitude
            cardPrice = df.loc[i,'Price']
            cardBedrooms = df.loc[i,'Bedrooms']
            cardBaths = df.loc[i,'Baths']
            cardArea = df.loc[i,'Area']
            temp = pd.DataFrame([[cardPrice, cardAddress_x, cardAddress_y, cardBedrooms, cardBaths, cardArea]],columns=["Price", "Address_x","Address_y", "Bedrooms", "Baths","Area"])
            df2 = df2.append(temp)
        else:
            list = df.loc[i,'Address'].split(",")
            length =len(list)
            addressArea=list[length-2]
            print("second last array element ",list[length-2])
            location=geolocator.geocode(addressArea)
            if location:
                cardAddress_x = location.latitude
                cardAddress_y = location.longitude
                cardPrice = df.loc[i,'Price']
                cardBedrooms = df.loc[i,'Bedrooms']
                cardBaths = df.loc[i,'Baths']
                cardArea = df.loc[i,'Area']
                temp = pd.DataFrame([[cardPrice, cardAddress_x, cardAddress_y, cardBedrooms, cardBaths, cardArea]],columns=["Price", "Address_x","Address_y", "Bedrooms", "Baths","Area"])
                df2 = df2.append(temp)
            else:
                print("--ADDRESS LOCATION NOT FOUND",df.loc[i,'Address'])
                latitude = 0
                longitude = 0
                cardAddress_x = latitude
                cardAddress_y = longitude
                cardPrice = df.loc[i,'Price']
                cardBedrooms = df.loc[i,'Bedrooms']
                cardBaths = df.loc[i,'Baths']
                cardArea = df.loc[i,'Area']
                temp = pd.DataFrame([[cardPrice, cardAddress_x, cardAddress_y, cardBedrooms, cardBaths, cardArea]],columns=["Price", "Address_x","Address_y", "Bedrooms", "Baths","Area"])
                df2 = df2.append(temp)

print(df2)
df2.to_csv('House2.csv',index = False,header = True)            
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv('House2.csv')
df3 = pd.DataFrame(columns=["Price", "Address_x","Address_y", "Bedrooms", "Baths","Area"])
#columns=["Price", "Address_x","Address_y", "Bedrooms", "Baths","Area"]

for i in range(0,df.shape[0]):
    if df.loc[i,'Price'] == None or df.loc[i,'Price'] < 1000 or np.isnan(df.loc[i,'Price']):
        pass
    elif df.loc[i,'Address_x'] == 0.0:
        pass
    else:    
        df3 = df3.append(df.loc[i])

print(df3)
Area_avg = int(np.mean(df3.loc[:,'Area']) + 0.5)
Baths_avg = int(np.mean(df3.loc[:,'Baths']) + 0.5)
Bedrooms_avg = int(np.mean(df3.loc[:,'Bedrooms']) + 0.5)

df3['Area'] = df3['Area'].fillna(Area_avg)
df3['Bedrooms'] = df3['Bedrooms'].fillna(Bedrooms_avg)
df3['Baths'] = df3['Baths'].fillna(Baths_avg)

print(df3)
#df3.to_csv('House3.csv',index = False,header = True)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.dummy import DummyRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
```

```python
X1 = df3.iloc[1:,1]
X2 = df3.iloc[1:,2] 
X3 = df3.iloc[1:,3]
X4 = df3.iloc[1:,4]
X5 = df3.iloc[1:,5]
X = np.column_stack((X1,X2,X3,X4,X5))

y = df3.iloc[1:,0]
```

```python
mean_error = []
std_error=[]
order_range = [1, 2, 3, 4]
for order in order_range:
    X_poly = PolynomialFeatures(order).fit_transform(X)
    model = RandomForestRegressor(n_estimators = 100,random_state = 10)
    scores = cross_val_score(model, X_poly, y, cv = 10, scoring = 'neg_mean_squared_error')
    mean_error.append(np.array(scores).mean())
    std_error.append(np.array(scores).std())

plt.errorbar(order_range,mean_error,yerr = std_error) 
plt.title("Mean and standard deviation of the mean square error vs Order.")
plt.xlabel('Order')
plt.ylabel('Mean_Square_Error') 
plt.xlim((0.5,4.5))
plt.show()
```

```python
Tree_number = [25, 30, 40, 50, 60, 70, 80, 90, 100]
accuracy = []
std_error = []
oob_score = []

for tree_number in Tree_number:
    clf2 = RandomForestRegressor(n_estimators = tree_number, random_state = 10)
    scores = cross_val_score(clf2, X, y, cv = 10, scoring = 'neg_mean_squared_error')
    accuracy.append(np.array(scores).mean())
    std_error.append(np.array(scores).std())

for i in range(0, len(Tree_number)):
    print("Tree Number: " + str(Tree_number[i]) + ", Mean_Square_Error: " + str(accuracy[i]) + ", std: " + str(std_error[i]))

plt.errorbar(Tree_number,accuracy,yerr = std_error) 
plt.title("Mean and standard deviation of the mean square error vs Tree number.")
plt.xlabel('Tree Number')
plt.ylabel('Mean_Square_Error') 
plt.xlim((20,105))
plt.show()
```

```python
max_depth = [7,9,11,13,15,17,19]
min_samples_split = [15, 25, 35 ,50, 75]
accuracy = []
std_error = []

for i in range(0, len(max_depth)):
    for k in range(0,len(min_samples_split)):
        clf2 = RandomForestRegressor(n_estimators = 90, random_state = 10, 
                                     max_depth = max_depth[i],
                                     min_samples_split = min_samples_split[k])
        scores = cross_val_score(clf2, X, y, cv = 10, scoring = 'neg_mean_squared_error')
        accuracy.append(np.array(scores).mean())
        std_error.append(np.array(scores).std())
        
        print("Max_depth: " + str(max_depth[i]) + 
              ", Min_samples_split: " + str(min_samples_split[k]) + 
              ", Mean_Square_Error: " + str(accuracy[k]) + 
              ", std: " + str(std_error[k]))
        
    plt.errorbar(min_samples_split, accuracy, yerr = std_error) 
    plt.title("Mean and standard deviation of the mean square error vs min_samples_split when max_depth = " + str(max_depth))
    plt.xlabel('min_samples_split')
    plt.ylabel('Mean_Square_Error') 
    plt.xlim((10,80))
    plt.show()
    
    accuracy.clear()
    std_error.clear()
```

```python
min_samples_leaf = [1,3, 5, 10, 20, 30]
min_samples_split = [2, 5, 15, 25, 35]
accuracy = []
std_error = []

for i in range(0, len(min_samples_leaf)):
    for k in range(0,len(min_samples_split)):
        clf2 = RandomForestRegressor(n_estimators = 90, random_state = 10, 
                                     max_depth = 15,
                                     min_samples_split = min_samples_split[k],
                                     min_samples_leaf = min_samples_leaf[i])
        scores = cross_val_score(clf2, X, y, cv = 10, scoring = 'neg_mean_squared_error')
        accuracy.append(np.array(scores).mean())
        std_error.append(np.array(scores).std())
        
        print("Min_samples_leaf: " + str(min_samples_leaf[i]) + 
              ", Min_samples_split: " + str(min_samples_split[k]) + 
              ", Mean_Square_Error: " + str(accuracy[k]) + 
              ", std: " + str(std_error[k]))
        
    plt.errorbar(min_samples_split, accuracy, yerr = std_error) 
    plt.title("Mean and standard deviation of the mean square error vs min_samples_split when min_samples_leaf = " + str(min_samples_leaf))
    plt.xlabel('min_samples_split')
    plt.ylabel('Mean_Square_Error') 
    plt.xlim((0,40))
    plt.show()
    
    accuracy.clear()
    std_error.clear()
```

```python
max_features = [1, 2, 3, 4, 5]
accuracy = []
std_error = []

for i in range(0,len(max_features)):
    clf2 = RandomForestRegressor(n_estimators = tree_number, random_state = 10, 
                                 max_depth = 15, 
                                 min_samples_split = 2, 
                                 min_samples_leaf = 1, 
                                 max_features = max_features[i])
    scores = cross_val_score(clf2, X, y, cv = 10, scoring = 'neg_mean_squared_error')
    accuracy.append(np.array(scores).mean())
    std_error.append(np.array(scores).std())

for i in range(0, len(max_features)):
    print("Max_features: " + str(max_features[i]) + ", Mean_Square_Error: " + str(accuracy[i]) + ", std: " + str(std_error[i]))

plt.errorbar(max_features, accuracy, yerr = std_error) 
plt.title("Mean and standard deviation of the mean square error vs max_features.")
plt.xlabel('max_features')
plt.ylabel('Mean_Square_Error') 
plt.xlim((0,6))
plt.show()
```
